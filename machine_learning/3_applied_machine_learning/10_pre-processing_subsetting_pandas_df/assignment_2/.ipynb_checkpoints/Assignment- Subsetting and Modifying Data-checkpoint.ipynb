#### `READ THE BIG MART SALES DATA`


---

```python
import pandas as pd
```

```python
data = pd.read_csv('dataset/big_mart_sales.csv')
```

```python
data.head()
```

---

#### `Select the rows where Item_MRP is less than 100`


---

```python
# WRITE YOUR CODE HERE
def mrp_less_100(mrp):
    if mrp < 100:
        return mrp
```

```python
data['Item_MRP_less_100'] = data.Item_MRP.apply(lambda x: mrp_less_100(x))
```

```python
data.loc[data.Item_MRP_less_100.isna() == False]
```

---

#### `Select the rows where Outlet_Location_Type is Tier 1 and Outlet_Establishment_Year is from the list [1987, 1988, 2009]`

--- 

```python
data.loc[(data.Outlet_Location_Type == 'Tier 1') & (data.Outlet_Establishment_Year.isin([1987, 1988, 2009]))]
```

#### `Update the values of the column Outlet_Type using the mapping given below` 


---

```python
mapping = {
    'Supermarket Type1' : 1,
    'Supermarket Type2' : 2,
    'Grocery Store'     : 3,
    'Supermarket Type3' : 4
}
```

```python
# WRITE YOUR CODE HERE
data['Outlet_Type'] = data.Outlet_Type.map(mapping)
data
```
