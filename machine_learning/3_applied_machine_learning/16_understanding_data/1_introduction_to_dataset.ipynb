#### Look at the available features

Data source 1 (Customer)    ---> Data extraction and collection
Data source 2 (Transaction) ---> Data extraction and collection
Data source 3 (Rank Branch) ---> Data extraction and collection

**Customer data source has following features:-**
1. customer_id - Customer Id
2. vintage - Association of the customer with the bank in number of days
3. age - Age of customer
4. gender - Gender of customer
5. dependents - Number of dependents
6. occupation - Occupation of the customer
7. city - City of the customer (anonymised)


**Transaction data source has following features:-**
1. current_balance - balance as of today
2. current_month_credit - Total credit amount current month
3. current_month_debit - Total debit amount current month
4. current_month_balance - Average balance of current month
5. previous_month_end_balance - End of month balance of previous month
6. previous_month_credit - Total credit amount previous month
7. previous_month_debit - Total debit amount previous month
8. previous_month_balance - Average balance of previous month
9. average_month_balance_prevQ - Avg balance in previous quarter
10. average_monthly_balance_prevQ2 - Avg balance in previous to previous quarter
11. customer_nw_category - Net worth of customer (3: Low, 2:Medium, 1: High)


**Bank branch data source has following features:-**
1. branch_code - Branch code for customer account
2. last_transaction - Day the last transaction was done
