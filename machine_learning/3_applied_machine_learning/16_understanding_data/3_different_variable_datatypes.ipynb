#### Different Variable Data Types


* Data Types
    * Categorical
        * Nominal
        * Ordinal
    * Numerical
        * Discrete
        * Continuous
        
#### Categorical Data Type
* Also known as qualitative data
* Represent characteristics or qualities
* Generally, airthmetic operation cannot be perform on it.
E.g. Your gender? 
Position of rider in horse racing


**1. Nominal Data**
    * represent Discrete unit and are used to label variables that have no quantitative value. Just like labels
    * No specific order
    * Changing label doesn't impact meaning
    e.g. Male and Female in gender can be represented in one or zero.

**2. Ordinal Data**
    * Discrete
    * Specific Order
    * Labels represent meaning
    e.g. Grades like A+ is better than A and A is better than B or so on.
    
**Nominal vs Ordinal Data types**

| Nominal | Ordinal |

| No quantitative meaning | Some quantitative meaning |

**1. Discrete Data**
* Takes only certain values
* Can't be measure but it can counted

e.g. Your sleeping hours?


**2. Continuous Data**
* Can take any value
* Cannot be counted but measured

Ex - When did you sleep last night?

The answer can take any value such as Hours, minutes, seconds, blank seconds and so on.
