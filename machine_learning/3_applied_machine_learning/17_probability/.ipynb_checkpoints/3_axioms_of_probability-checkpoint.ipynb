#### Axioms of Probability

1. Probability of any event E, 0 <= P(E) <= 1 where 1 indicates definite outcome of an event and 0 indicates no outcome
2. Probability for sample space, P(S) = 1
3. P(A union B) = P(A) + P(B) for mutually disjoint events


#### Axiom 1: Probability of Event
* For any event E, 0 <= P(E) <= 1

P(event) = count of outcomes in event / count of outcomes in sample space

#### Question 2
Mutually exhaustive events can not be mutually exclusive events.

Consider U = {1, 2, 3, 4, 5}

A = {1, 2, 3} and B = {4, 5}

A and B are mutual exclusive and mutual exhaustive.

A = {1, 2, 3} and B = {3, 4, 5}

Mutually exhaustive events can not be mutually exclusive events. It is not always True therefore false.
