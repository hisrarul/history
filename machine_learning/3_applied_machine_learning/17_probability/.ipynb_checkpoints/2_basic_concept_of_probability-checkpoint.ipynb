#### Basic concept of Probability
* Complement
* Intersection
* Union
* Additive rule

#### Complement
The complement of an event E in a sample space S, is the collection of all outcomes in S are not element of the set E.

E.g.

Event E: even number on rolling dice.

Event E(complement): odd number on rolling dice.

P(Event(complement)) = 1 - P(E)

Ex:- Dataset from churn, Probability of customer being old?
Sample space = {Young, Old}

young = 12
old = 3
sample space = 12 + 3 = 15

Consider event = old
event(complement) = young

P(E) = 1 - P(event(complement))
     = 1 - 12/15
     = 3/15
     = 1/5
     

#### Intersection

Among all the customers who are not going to churn, probability of one being a self employed male person?

From churn dataset, we know

**event A**
male not churn = 9
self employed male not churn = 5

**event B**
self employed not churn = 6
self employed male not churn = 5

Total retaining customers = 10
P(A intersection B) = 5/10

#### Union
Union (U) is the collection of all outcomes that are elements of any of the events.

Ex- Among all the customers going to churn, one being a female or a self employed person?

**event A**
female churn = 3

**event B**
self employed churn = 2

self employed female churn = 1

total customer reducing interaction = 5
P ( A union B) = 4/5 = 0.8

#### Additive Rule

Additive rule of Probability = P(A U B) = P(A) + P(B) - P (A intersection B)

#### Example 1:
What is the probability that atleast one tail will appear in four tosses of a fair coin.

Probability of heads, P(H) = 1/2 x 1/2 x 1/2 x 1/2 = 1/16

Probability of atleast 1 tail = 1 - P(H) = 1 - 1/16 = 15/16


#### Example 2:

What is the probability that in fair dice, the number rolled is both odd and greater than three.

The only number which is odd greater than 3 in a dice is 5

so probability of getting 5 in a fair dice will be 1/6.

#### Example 3:
What is the probability that in fair dice, the number rolled is either odd or greater than three.

Number which are odd = 1, 3, 5

Number which are greater than 3 = 4,5,6

Probability that in fair dice, the number rolled is either odd or greater than three will be number which are odd + number which are greater than 3

so probability will be getting numbers = {1,3,4,5,6}

probability = 5/6


#### Example 4:

In an organization of 70 employees, what is the probability that at least two employees will have birthday on the same data?

Complementation Law:

P(A)c = 1 - P(A).P(A)c = 1-P(A)

using complementation law in birthday paradox

P(atleast 2 have same birthday) = 1 - P(no one birthday)

Calculation of probability of no 2 having same birthday

P(1B) = 365/365

P(2B) = 364/365

P(3B) = 363/365

...

P(70B) = 365-69/365

since, P(no birthday) = (365/365)(364/365)(363/365)....(365-69/365)

Therefore, P(2Birthday) = 1 - P(noBirthday)

Beautify, the equation with factorial!

P(2Birthday) = 1 - (365!/(365^70)(365-70!) = 0.9999
