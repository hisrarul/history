#### Random experiment
* Random experiment is a process with a number of possible outcomes. e.g. tossing a coin, rolling a die, fetching a card from a pack, etc
* Those outcomes are not necessarily certain.
* It can be repeated numerous times under the same conditions.
* Generally, the outcome of an individual random experiment is independent and identically distributed.

**In our churn example,** the profession of a customer has only three categories. Ex Self employed, Salaried, Retired.


#### Sample Space
Sample space associated with a random experiment is a set of all possible outcomes.

Ex - The profession of a cutomer are categories into Self employed, Salaried, Retired.

Sample space = Self employed, Salaried, Retired


#### Event
An event is a subset of sample space.

Ex 1:- If customer is currently working.

Event = {salaried, self employed}


Ex 2:- 
sample space of a rolling die = 1, 2, 3, 4, 5, 6
event 1: Getting 3 on a dice
single outcome meaning event = {3}

event 2: Getting odd number on a dice
multiple outcomes = {1, 3, 5}

Both of the events will have **6** as sample space.

#### Probability

Probability is the likelihood or chance of an event occuring.

The probability of an Event E, is a number P, between 0 and 1 that measures the likelihood that Event will occur.

P(event) = count of outcomes in event / count of outcomes in sample space

**P = 0  ---> Impossible event**
Ex- Getting 8 on rolling a six faced dice

**P = 1  ---> Certain event**
Ex- Getting a number less than 7, on rolling a six faced dice


#### Random experiment example
**Probability of the churn example**
let's take 15 customer from churn example dataset

sample sample = {salaried, self employed, retired}
salaried = 5
self employed = 9
retired = 1

count of outcomes in sample space = 5 + 9 + 1 = 15

P(Customer = Salaried) = 5/15= 0.333
P(Customer = Self employed) = 9/15 = 0.6
P(Customer = Retired) = 1/15 = 0.067


#### Definite and impossible outcome

Hence, from the above description, we can say that

Salaried Female = 1  ---1 (Churn)--> Definite outcome
