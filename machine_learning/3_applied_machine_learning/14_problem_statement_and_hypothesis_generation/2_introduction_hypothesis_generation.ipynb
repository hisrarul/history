#### What is hypothesis?

A hypothesis is a possible view or assertion of a data scientist about the problem he or she is working on. It may or may not be true.

#### Why do we need hypothesis generation?

| Non-hypothesis driven data analysis | Hypothesis driven analysis |
|----|-----|
| Collect the entire data | List hypothesis before looking at data |
| Analyze the every variable available | Smaller & specific pieces to work with |
| Could lead to narrow vision | Provides an UNBIASED view |
| | Exhaustive set ---> Collect data ---> Test each hypothesis |
