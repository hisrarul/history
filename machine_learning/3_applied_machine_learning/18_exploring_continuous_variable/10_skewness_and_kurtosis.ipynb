#### Skewness

* skewness can be calculated with pandas.DataFrame.skew
* -0.5 to 0.5 ---> Symmetric
* Less than -0.5 ---> Negatively skewed
* Greater than 0.5 ---> Positively skewed

#### Kurtosis
* Kurtosis gives an idea of the tails of a distribution.
    * Symmetric
        * K=0 (Mesokurtic)
        * Tails similar to normal distribution
    * High presence of extreme values
        * K>0 (Leptokurtic)
        * Longer distribution, fatter tail
    * Less Presence of extreme values means lesser data points lie along the tail
        * K<0 (Platykurtic)
        * Shorter distribution, thinner tail
