#### Probability Density Function

P(event) = count of outcomes in Event / count of outcomes in Sample space

* For continuous variable ?
* Sample space has infinite values e.g. between 1 and 2 there will be lot of decimal values and these will comprise the sample space.
* So probability at a particular value is 0
* That's why we will probability density function for continuous variable.

#### Types of continuous distribution

**Uniform distribution**: The uniform distribution is a continuous probability distribution and is concerned with events that are equally likely to occur.

The probability density function for uniform distribution is 

f(x) = 1 / b - a  if b >= x >= a

f(x) = 0 otherwise
