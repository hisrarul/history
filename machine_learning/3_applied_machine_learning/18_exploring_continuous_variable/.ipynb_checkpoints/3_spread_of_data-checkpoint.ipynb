#### Spread of Data

The measure of central tendency are not adequate to describe the data.

e.g. var1 = -2, -1, 0, 1, 2

var2 = -4, -3, 0, 3, 4

The mean and median of both are zero but are they really the same. This is where spread of data comes into the picture.

Measures of spread describe how similar or varied the set of data points are.

Spread also known as variability and dispersion.

#### Variance and Standard Deviation
* Variance is the average squared difference of the values from the mean.
* **Standard deviation** is the square root of variance.
* If not squared positive deviation can cancel the negative deviation so taking square of difference actually helps.
