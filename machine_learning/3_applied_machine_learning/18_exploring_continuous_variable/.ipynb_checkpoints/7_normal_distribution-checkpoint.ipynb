#### Properties of Normal Distribution
* Also known as bell shaped or gaussian distribution
* Symmetric means left side distribution mirrors the right distribution
* Mean, mode, median all are equal.
* Area under the curve = 1

#### Empirical rule for normal distribution

Empirical rule for normal distribution: 68-95-99.7 rule.

68.27% of data lies within one standard deviation of the mean.

95.45% of data within two standard deviation of the mean.

99.73% of data within three standard deviation of the mean.

As shown in the figure

#### Standard Normal Distribution

when u=0 and sigma=1

Z = x - u / sigma

**Example**:
Ross 65 marks in paleontology

Rachel, 80 marks in fashion designing

did rachel perform better than Ross?

Can't say. Direct comparison of scores will not work.

Suppose paleontology marks follows normal distribution with mean=60 and standard deviation = 4

Fashion designing marks, follows normal distribution with mean(u) = 79 and standard deviation = 2

Calc Z for Ross = 65 - 60 / 4 = 1.25

Calc Z for Rachel = 80 - 79 / 2 = 0.5

In reality Ross perform better than Rachel.
