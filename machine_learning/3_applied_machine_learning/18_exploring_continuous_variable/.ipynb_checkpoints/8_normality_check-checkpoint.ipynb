#### Normality Check

Why Normality check?

* There are many machine learning methods which performs better on the data which is normally distributed.

* There are many statistical test consider the assumption of normality so we often do transformations on variables to convert them into normal and later use normality check to confirm it work.

* It helps in outlier detection.

#### Normality Check: Histogram

**Visuality Check**
Frequency against bins to get the histogram.

#### Normality Check: KDE

**Visuality Check**
KDE is a method to estimate pdf of data generating distribution.


#### Normality Check: Q_Q Plot
Quantile, divide the probability distribution or observations in equal intervals.

* 2 quantile - Median
* 4 quantile - Quartile
* 10 quantile - Decile
* 100 quantile - Percentile

**Visuality check**:
Q-Q plots take theoretical normal distribution quantiles and compare them against observed data quantiles.

The scatter plot will roughly form a straight line.

**Example**:

If the following data comes from a normal distribution?

106.75, 101.57, 100.78, 89.67, 120.08, 106.93, 125.61, 111.8, 101.51

Step 1: Sort values

89.67, 100.78, 101.51, 101.57, 106.75, 106.93, 111.8, 120.08, 125.61

Step 2: Divide a normal curve into n+1 (here 10) equally-sized areas.

Step 3: Find the z value for each of the segment i.e. 10%, 20%, 30% .... 100%

step 4: Plot the sorted data values against observed z value.
