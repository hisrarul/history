#### Central tendency

It is the way of summarizing the data present at hand by looking some sort of central value. Central tendency provides us the descriptive summary of a variable.

* Descriptive summary

* Centrality of the data

* Measures for central tendency
    * Mean
    * Median
    * Mode

#### Mean
* Sum of all data points divided by the total number of data points.
* The mean value of a dataset is the airthmetic avearage value.
* Every data point is used in its calculation.
* More useful for continuous data points.
* Mean is a good indicator of central tendency for continuous data points without extreme points.

### Median
* Median is the middle value. It is the value that splits the dataset in half.
* Median is a good indicator for ordered data points or continous data points with extreme points.

e.g. sal = 15K, 14k, 12k, 15k

avg sal = 15+14+12+15/4 = 56/4

if high value added then avg will be impacted

sal = 15K, 14k, 12k, 15k, 200k

avg sal = 15+14+12+15+200/5 = 256/5

Hence, midpoint is taken **after sorting** then avg is counted.

for odd = X(n+1/2) 

for even = X(n+1/2) + X(n/2) / 2
