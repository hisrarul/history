## Skewed Distribution


#### Symmetric Distribution

* Symmetric distribution if the tail on both side of the distribution are identical.
* The mean, mode and median will be very close to each other.
* Mode is the most frequent value in the dataset.
* But how mode calculated for continuous variables.

#### Mode for continuous variable
The mode of a continuous random variable corresponds to the value at which the probability density function reaches maximum, or a peak.

#### Right (Positive) Skewed

* Positive skewness if the tail on the right side of the distribution is longer
* The mean and median will be greater than the mode.


#### Left (Negative) Skewed
* Negative skewness if the tail on the left side of the distribution is longer
* The mean and median will be lesser than the mode.
