#### Introduction to Predictive Modeling

*What is Predictive Modeling?*
1. Making use of past data and attributes
2. Predict the future using this data.


Predictive which of the below is the example of predictive modeling
* Gender, location, age, past movies -> Predictive modeling
* Factors responsible for sales reduction -> not a predictive modeling bcz
    * Analysing past data
    * No future outcome or forecast 
* Viewing a website today's traffic -> not a predictive modeling bcz
    * No past data involved
    * No future outcome required
* Predicting stock price movement -> predictive modeling bcz
    * Analysing past stock prices
    * Analysing similar stocks
    * Future stock price required

#### Predictive Modeling

It is a process to create statistical model for estimating/predicting the future behaviour based on past data.

It is used in several organizations at a large scale. 
1. Political parties used this technique to know their chances of winning on micro geographic areas.
2. Walmart and Tesco predict the next purchase of their customer.

Predictive modeling is used to predict future behaviour or results based on past data and trends.

Example: A retail bank wants to know the default behaviour of its credit card customers. They want to predict the probability of default for each customer within next 3 months.

Assuming

* probability of default would lie between 0 and 1.
* Assume 10% of customers will default in next 3 months.

This generailzation is not sufficient to do business decisions. Now, what do these predictive models do?
It moves the probability towards one of the two extremes based on attributes from past information. A customer with volatile income is more likely (closer to 1) to default. A customer with healthy credit history for last 10 years has low chances of default (closer to 0).


Probability of default for each customer in next 3 months = 0.1



#### Steps of Predictive modeling
Algorithm Selection -> Training Model -> Prediction/Scoring


#### Algorithm Selection
**Data exploration stage**
A dependent variable is what we measure in a model. The name itself says these variables are dependent on independent variables and **independent variables** are those which we used to measure or predict the dependent variables. Algorithm in predictive modeling can be broadly classified in two types:

1. Supervised Learning
2. Unsupervised Learning

Have dependent variable ---Y---> Supervised Learning
Have dependent variable ---N---> Unsupervised Learning

Example
If you want to predict wether the customer will default or not. It will be a using a supervised learning algorithm.
Why? We want to predict a dependent variable based on independent variables.

If your organization wants to segment it's customer so that this segmentation can be used across various engagement, you would be using unsupervised learning.

**Have dependent variable ---Y---> Supervised Learning ---> Is dependent variable continuous? ----> Regression**

Example: Estimate the total sales by an insurance agent in next 3 months

Algorithm:

* Simple Linear Regression
* Multiple Regression


**Have dependent variable ---Y---> Supervised Learning ---> Is dependent variable continuous? ----> Classification**

Example: Predict the customer will buy the product or not

Common Algorithms:

* Logistics regression
* Decision Tree
* Random Forest

**If the dependent variable is not present**
Then we will use unsupervised learning e.g. clustering

Have dependent variable ---N---> Unsupervised Learning ----> Clustering

Example:
Customer segmentation for defining right product.

Common Algorithms:
* K-means
* Hierarichal Clustering


#### Training Model
It is a process to learn relationship/correlation between independent and dependent variables.

* Dataset
    * Train
        1. Past data (Known dependent variable)
        2. Used to train model
    * Test
        1. Future data (Unknown dependent variable)
        2. Used to score
        
It is a process to estimate/predict dependent variable of test data set by applying model rules
