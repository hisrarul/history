#### Machine Learning Pipeline

Any machine learning projects involves a series of sequential steps known as machine learning pipeline.

#### Why do we need EDA

* EDA helps us to draw out the important inferences and insight from the data whereas insights help us in
    * Variable Identification
    * Behaviour of Variables
    * Relationship between Variables
* EDA also help us to check the consistency of the data. For e.g.
    * All data present?
    * Any missing values?
    * Any outliers?
* Feature engineering

All these three components acts as bedrock for predictive modeling.

#### Steps in EDA
1. Variable Identification
2. Univariate Analysis e.g. analysis for numerical data such as age and categorical data such as gender
3. Bivariate/multivariate Analysis i.e. Defines the relationship between different variable to find out how they depend on each other or how they vary with each other.
4. Missing Value Treatment i.e. why values are missing
5. Outlier Removal
